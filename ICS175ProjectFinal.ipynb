```
# Poor implementation to grab min/max x and y
processed_data = dict()
for key, value in data.items():
  min_x = sys.maxsize
  min_y = sys.maxsize
  max_x = -sys.maxsize -1
  max_y = -sys.maxsize -1
  for x in value:
    if(min_x > x[0]):
      min_x = x[0]
    if(min_y > x[1]):
      min_y = x[1]
    if(max_x < x[0]):
      max_x = x[0]
    if(max_y < x[1]):
      max_y = x[1]
  processed_data[key] = [[min_x, min_y], [max_x, max_y]]
  
```
